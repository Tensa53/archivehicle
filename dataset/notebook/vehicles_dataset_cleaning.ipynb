```python
# importo il csv del dataset come un dataframe pandas
import pandas as pd
df = pd.read_csv("../vehicles_dataset_original.csv")
df.head()
```

```python
# mostro le informazioni del dataset
df.info()
```

```python
# rimuovo l'anno dal nome del veicolo
df.name = df.name.str[5:]
df.head()
```

```python
# rimuovo le righe che presentano valori nulli
df = df.dropna()
df.head()
```

```python
# rimuovo caratteri di newline/spacing in eccesso
df.description = df.description.replace(r"^\s+|\s+$", r"", regex=True)
df.head()
```

```python
# cambio il tipo di cylinders e doors da float a int
df = df.astype({'cylinders': int, 'doors': int})
df.head()
```

```python
# recupero gli indici delle righe dove i valori numerici non sono significativi
dff = df[(df.price == 0.0) | (df.cylinders == 0) | (df.doors == 0)]
ln = dff.index
dff.head()
```

```python
# rimuovo le righe con gli indici precedentemente recuperati
df = df.drop(ln)
df.head()
```

```python
# recupero gli indici delle righe dove la colonna description presenta valori non validi
dff = df[(df.description == ".")]
lp = dff.index
dff.head()
```

```python
# rimuovo le righe con gli indici precedentemente recuperati
df = df.drop(lp)
df.head()
```

```python
# recupero gli indici delle colonne con le descrizione corte (minore di 100 caratteri)
dff = df[df['description'].str.len() < 100]
lc = dff.index
dff.head()
```

```python
# rimuovo le righe con gli indici precedentemente recuperati
df = df.drop(lc)
df.head()
```

```python
# rimuovo tutte le righe duplicate sulla base del nome del veicolo
df = df.drop_duplicates(subset=['name'])
df
```

```python
# visualizzo tutti i tipi di chassis, escludendo i duplicati
df.rename(columns = {'body': 'chassis'}, inplace=True)
df.chassis.drop_duplicates()
```

```python
# semplifico i tipi di chassis, raggruppando tutto in Van
df.chassis = df.chassis.replace("Cargo Van","Van")
df.chassis = df.chassis.replace("Passenger Van","Van")
df.chassis = df.chassis.replace("Minivan","Van")
df.head()
```
